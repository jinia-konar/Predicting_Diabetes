```
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_yaml
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Predicting Diabetes/train.csv")
#df.set_index('0', inplace=True)
df=df.drop(columns=['Unnamed: 0'])
print(df.head())
```

```
X = df.loc[:, df.columns!='Outcome']
Y = df.loc[:, 'Outcome']
```

```
model = Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
model.fit(X,Y,epochs=200)
```

```
scores = model.evaluate(X,Y)
print("Training Accuracy: %.2f%%\n" % (scores[1]*100))
```

```
model_json = model.to_json()
with open("/content/drive/My Drive/Colab Notebooks/Predicting Diabetes/model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("/content/drive/My Drive/Colab Notebooks/Predicting Diabetes/model.h5")
print("Saved model to disk")
```
